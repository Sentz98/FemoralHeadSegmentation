```python
import pandas as pd

# Load the data from the CSV file
data = pd.read_csv('data/data.csv')

# Display the first few rows of the data
data.head()
```

```python
import matplotlib.pyplot as plt

# Age Distribution using a histogram
plt.figure(figsize=(10, 4))
plt.hist(data['Age'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.grid(axis='y')
plt.show()
```

```python
# Gender Distribution using a pie chart
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(8, 4))
gender_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#ff9999'])
plt.title('Gender Distribution of Patients')
plt.ylabel('')
plt.show()
```

```python
# Pathology Distribution using a bar chart
pathology_counts = data['Type'].value_counts()
plt.figure(figsize=(10, 4))
pathology_counts.plot(kind='bar', color='#86bf91', edgecolor='k', alpha=0.7)
plt.title('Pathology Distribution of Patients')
plt.xlabel('Pathology Type')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
